```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'], y=[2,3,1])
plt.show()
```

```python
sns.set(style='darkgrid', palette='deep')
sns.barplot(x=['A','B','C'], y=[2,3,1])
plt.show()
```

```python
sns.set(style='whitegrid', palette='icefire', font="DejaVu Sans", font_scale=4)
sns.barplot(x=['A','B','C'], y=[2,3,1])
plt.show()
```
