```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_ylabel('Y')
axes[1].set_xlabel('X')
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x,y)
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_ylabel('Y')
axes[1].set_xlabel('X')
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x,y,label="sin(x)")
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend() ## coloca el "sin(x) que defin en la segunda lnea"
axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_ylabel('Y')
axes[1].set_xlabel('X')
```
