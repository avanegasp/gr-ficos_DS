```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'COLOMBIA', 'GERMANY', 'MEXICO']
population = [1000,3000,800,2500,1700]
```

```python
plt.bar(country,population)
plt.show()
```

```python
plt.bar(country,population,width=0.5, color=['red','yellow','black','green','orange'])
plt.show()
```

```python
plt.bar(country,population,width=0.5, color=['red','yellow','black','green','orange'])
plt.xticks(np.arange(5),('India', 'japon', 'mexico', 'colombia', 'alemania'),rotation=45)
plt.show()
```
