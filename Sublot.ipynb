```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(3,5,2)
plt.plot(x,y,'r--')

plt.subplot(3,5,3)
plt.hist(y)
plt.show()
```

```python
plt.subplot(5,2,3)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')

plt.subplot(5,3,3)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')

plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```
