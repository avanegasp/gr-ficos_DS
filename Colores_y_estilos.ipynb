```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('fivethirtyeight')
```

```python
fig, ax=plt.subplots(figsize=(8,6))
```

### modififcaciones de tipo matlab

```python
fig, ax=plt.subplots(figsize=(8,6))
ax.plot(x,x+1,'r')
ax.plot(x,x+2,'g')
ax.plot(x,x+3,'p')
ax.plot(x,x+4,'o')
```

```python
fig, ax=plt.subplots(figsize=(8,6))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'g:')
```

Pyplot

```python
fig, ax=plt.subplots(figsize=(8,6))
ax.plot(x,x+1,color='black',alpha=0.8, linewidth=5, linestyle='-',marker='x',markerfacecolor='#FF0000')
ax.plot(x,x+2,color='pink', linestyle='--',marker='o')
ax.plot(x,x+3,color='#66FF66', linestyle='dashed',marker='v')
ax.plot(x,x+4,color='yellow', linestyle=':',marker='P',markersize=10, markerfacecolor='#FF0000')
```
