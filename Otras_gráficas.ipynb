```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

```python
data
```

## distibuci√≥n es histograma

```python
plt.hist(data)
plt.show
```

```python
plt.hist(data,bins=30,width=2)
plt.show
```

```python
plt.hist(data,bins=20,histtype='step')
plt.show
```

```python
## boxplot gr√°fico de cajas y bigotes
```

```python
plt.boxplot(data)
plt.show
```

```python
plt.boxplot(data,vert=False)
plt.show
```

```python
plt.boxplot(data,vert=False, patch_artist=True)
plt.show
```

```python
plt.boxplot(data,vert=False, patch_artist=True,notch=True)
plt.show
```

```python
data = np.append(data,200)
plt.boxplot(data,vert=False, patch_artist=True,notch=True,showfliers=True)
plt.show
```

```python
data = np.append(data,200)
plt.boxplot(data,vert=False, patch_artist=True,notch=True,showfliers=False)
plt.show
```

### gr√°ficos de dispersi√≥n o sCATTER

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='o',alpha=0.5)
plt.show()
```
