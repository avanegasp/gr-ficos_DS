```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
sns.displot(data=tips)
```

```python
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=False)
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=True)
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=False, palette='dark')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=False, palette='dark', alpha=.25)
plt.show()
```
